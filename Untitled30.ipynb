```python
import warnings
warnings.filterwarnings("ignore")
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2 as cv
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,Dropout,BatchNormalization
from keras.preprocessing.image import ImageDataGenerator
```

```python
img_jasmine=cv.imread("Downloads/Rice_Image_Dataset/Rice_Image_Dataset/Jasmine/Jasmine (10005).jpg")
plt.imshow(cv.cvtColor(img_jasmine,cv.COLOR_BGR2RGB))
plt.axis('off')
print(img_jasmine.shape)
```

```python
train_gen=ImageDataGenerator(rescale=1./255,validation_split=0.3)
train_data=train_gen.flow_from_directory("Downloads/Rice_Image_Dataset/Rice_Image_Dataset",target_size=(100,100),batch_size=32,class_mode='categorical',shuffle=True,subset='training')
test_data=train_gen.flow_from_directory("Downloads/Rice_Image_Dataset/Rice_Image_Dataset",target_size=(100,100),batch_size=1,shuffle=False,subset='validation')
```

```python
train_data.class_indices
```

```python
model=Sequential()
model.add(Conv2D(filters=32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(100,100,3)))
model.add(MaxPooling2D(pool_size=(2,2),strides=2))
model.add(Flatten())
model.add(Dense(46,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(5,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit_generator(train_data,epochs=10,validation_data=test_data,callbacks=tf.keras.callbacks.EarlyStopping(monitor="val_loss",patience=3))
```

```python
y_predictions = model.predict(test_data)
```

```python
y_pred= np.argmax(y_predictions,axis=1)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_data.classes,y_pred)
sns.heatmap(cm, annot=True)
```
